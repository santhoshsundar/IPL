# Exploratory Analysis of IPL Data

## Introduction
We have complete Indian Premier League data from 2008 to 2017. When this volume of ipl data is made available, only justice that can be done is to draw cool insights and get to know our players better.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
```

```python
matches = pd.read_csv('../IPL/ipl/matches.csv')
deliveries = pd.read_csv('../IPL/ipl/deliveries.csv')
```

## Number of Total Matches Played

```python
num_matches = matches['team1'].count()
num_matches
```

### Preprocessing our data

```python
matches.drop(['umpire3'],axis=1,inplace=True)
deliveries.fillna(0,inplace=True)
```

```python
matches.replace(['Mumbai Indians','Kolkata Knight Riders','Royal Challengers Bangalore','Deccan Chargers','Chennai Super Kings',
                 'Rajasthan Royals','Delhi Daredevils','Gujarat Lions','Kings XI Punjab',
                 'Sunrisers Hyderabad','Rising Pune Supergiants','Kochi Tuskers Kerala','Pune Warriors','Rising Pune Supergiant']
                ,['MI','KKR','RCB','DC','CSK','RR','DD','GL','KXIP','SRH','RPS','KTK','PW','RPS'],inplace=True)

deliveries.replace(['Mumbai Indians','Kolkata Knight Riders','Royal Challengers Bangalore','Deccan Chargers','Chennai Super Kings',
                 'Rajasthan Royals','Delhi Daredevils','Gujarat Lions','Kings XI Punjab',
                 'Sunrisers Hyderabad','Rising Pune Supergiants','Kochi Tuskers Kerala','Pune Warriors','Rising Pune Supergiant']
                ,['MI','KKR','RCB','DC','CSK','RR','DD','GL','KXIP','SRH','RPS','KTK','PW','RPS'],inplace=True)
```

### Number of matches won by each teams

```python
match_winner = matches['winner'].value_counts()
data = [go.Bar(
            x=match_winner.index.values.tolist(),
            y=match_winner.values.tolist(),
            text = match_winner.values.tolist(),
            textposition = 'auto',
    )]
layout = go.Layout(
    title='Number of Matches Won By Teams'
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

### Number of matches played by each teams

```python
matches_played_byteams = matches['team1'].value_counts() + matches['team2'].value_counts()
data = [go.Bar(
            x = matches_played_byteams.index.values.tolist(),
            y =  matches_played_byteams.values.tolist(),
            text = matches_played_byteams.values.tolist(),
            textposition = 'auto'
    )]
layout = go.Layout(
    title='Number of Matches Played By Teams of All Times'
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

### Number of matches played at different stadiums

```python
matches_played_stadium = matches['venue'].value_counts()
data = [go.Bar(
            x = matches_played_stadium.index.values.tolist(),
            y =  matches_played_stadium.values.tolist(),
            text = matches_played_stadium.values.tolist(),
            textposition = 'auto'
    )]

layout = go.Layout(
    title='Number of Matches Per Stadium'
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

### Win percentage of each teams

```python
matches_won_byteams = matches['winner'].value_counts()
win_percentage_byteams = 100 * (matches_won_byteams / matches_played_byteams)
win_percentage_byteams = win_percentage_byteams.round(1)
win_percentage_byteams = win_percentage_byteams.astype(str)
for i in range(13):
    win_percentage_byteams[i] = win_percentage_byteams[i] + "%"
win_percentage_byteams
```

```python
trace1 = go.Bar(
            x = matches_played_byteams.index.values.tolist(),
            y =  matches_played_byteams.values.tolist(),
            text = matches_played_byteams.values.tolist(),
            textposition = 'auto',name='Mathces Played')
trace2 = go.Bar(
            x = matches_won_byteams.index.values.tolist(),
            y =  matches_won_byteams.values.tolist(),
            text = matches_won_byteams.values.tolist(),
            textposition = 'auto',name='Matches Won')
trace3 = go.Bar(
            x = win_percentage_byteams.index.values.tolist(),
            y =  win_percentage_byteams.values.tolist(),
            text = win_percentage_byteams.values.tolist(),
            textposition = 'auto',name='Win Percentage')

data = [trace1, trace2, trace3]
layout = go.Layout(
    barmode='stack',
    title='Win Percentage'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

### Merge matches and deliveries

```python
matches_deliveries = matches[['id','season']].merge(deliveries, left_on = 'id', right_on = 'match_id', how = 'left').drop('id', axis = 1)
matches_deliveries.set_index(['season'],inplace=True)
matches_deliveries.sort_index(inplace=True,ascending=True)
```
